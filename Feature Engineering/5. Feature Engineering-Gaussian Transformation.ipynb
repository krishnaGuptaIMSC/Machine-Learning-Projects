## Gaussian Transformation

##### Convert distribution to Gaussian

### Guassian Transformation

1. Logarithmic Transformation
   
2. Reciprocal Trnasformation
  
3. Square Root Transformation
  
4. Exponential Trnasformation
  
5. Box Cox Transformation

```python
import pandas as pd

df=pd.read_csv('titanic.csv',usecols=['Age','Fare','Survived'])
df.head()
```

```python
### fillnan
df['Age']=df['Age'].fillna(df['Age'].median())
df.isnull().sum()
```

```python
import scipy.stats as stat
import pylab 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

#### If you want to check whether feature is guassian or normal distributed
#### Q-Q plot

def plot_data(df,feature):
    plt.figure(figsize=(15,6))
    plt.subplot(1,3,1)
    df[feature].hist(bins=35)
    plt.subplot(1,3,2)
    sns.distplot(df[feature])
    plt.subplot(1,3,3)
    stat.probplot(df[feature],dist='norm',plot=pylab)
    plt.show()
```

```python
plot_data(df,'Fare')
```

```python
plot_data(df,'Age')
```

### 1. Logarithmic Transformation

```python
import numpy as np

df['Fare_log']=np.log1p(df['Fare'])
plot_data(df,'Fare_log')
```

```python
df['Age_log']=np.log1p(df['Age'])
plot_data(df,'Age_log')
```

### 2. Reciprocal Trnasformation

```python
import numpy as np

df['Fare_reciprocal'] = np.reciprocal(df['Fare']+1)
plot_data(df,'Fare_reciprocal')
```

```python
df['Age_reciprocal'] = np.reciprocal(df['Age']+1)
plot_data(df,'Age_reciprocal')
```

### 3. Square Root Transformation

```python
df['Fare_sqaure']=df['Fare']**(1/2)
plot_data(df,'Fare_sqaure')
```

```python
df['Age_sqaure']=df['Age']**(1/2)
plot_data(df,'Age_sqaure')
```

#### 4. Exponential Transdormation


```python
df['Fare_exponential']=df.Fare**(1/1.2)
plot_data(df,'Fare_exponential')
```

```python
df['Age_exponential']=df.Age**(1/1.2)
plot_data(df,'Age_exponential')
```

### 5. BoxCOx Transformation
The Box-Cox transformation is defined as: 

T(Y)=(Y exp()1)/

where Y is the response variable and  is the transformation parameter.  varies from -5 to 5. In the transformation, all values of   are considered and the optimal value for a given variable is selected.

```python
df['Fare_Boxcox'],parameters = stat.boxcox(df['Fare']+1)
print(parameters)
plot_data(df,'Fare_Boxcox')
```

```python
df['Age_Boxcox'],parameters1 = stat.boxcox(df['Age']+1)
print(parameters1)
plot_data(df,'Age_Boxcox')
```
